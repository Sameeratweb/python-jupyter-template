### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Step 1: Define the data
categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

# Step 2: Create the bar plot with different colors for each bar
colors = ['red', 'blue', 'green', 'purple']  # Different colors for each bar
plt.bar(categories, values, color=colors)

# Step 3: Add a title to the plot
plt.title('Bar Plot: Categories vs Values')

# Step 4: Display the plot
plt.show()
```
